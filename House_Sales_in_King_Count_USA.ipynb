### Question  6

Fit a linear regression model to predict the <code>'price'</code> using the feature <code>'sqft_living'</code> then calculate the R^2. Take a screenshot of your code and the value of the R^2. You will need to submit it for the final project.


```python
#Enter Your Code, Execute and take the Screenshot
print("\n--- Fitting Linear Regression: sqft_living vs. Price ---")

# Create a LinearRegression object
lm = LinearRegression()

# Prepare the data
# We need to use 'sqft_living' (feature) and 'price' (target)
# The dataset 'kc_house_data_NaN.csv' implies missing values.
# We must drop any rows that have NaN in either of these critical columns.
model_data = df[['sqft_living', 'price']].dropna()

# Define Feature (X) and Target (Y)
# X must be a 2D array (DataFrame), Y can be a 1D array (Series)
X = model_data[['sqft_living']]
Y = model_data['price']

# Check if data is available after dropping NaNs
if X.empty or Y.empty:
    print("Error: No data available for 'sqft_living' and 'price' after dropping NaNs.")
else:
    # Fit the linear regression model
    lm.fit(X, Y)
    
    # Calculate and print the R^2
    # The score() method returns the coefficient of determination R^2 of the prediction.
    r_squared = lm.score(X, Y)
    print(f"The R-squared (R^2) value for the model (price ~ sqft_living) is: {r_squared}")
```

### Question 8

Use the list to create a pipeline object to predict the 'price', fit the object using the features in the list <code>features</code>, and calculate the R^2. Take a screenshot of your code and the value of the R^2. You will need to submit it for the final project.


```python
#Enter Your Code, Execut# --- Question 8: Pipeline with Scaling, Polynomial Features, and Linear Regression ---
print("\n--- Creating and Fitting a Data Pipeline ---")

# Create the list of tuples for the pipeline
Input = [('scale', StandardScaler()),
         ('polynomial', PolynomialFeatures(include_bias=False)),
         ('model', LinearRegression())]

# Create the pipeline object
# This pipeline will first scale the data, then create polynomial features,
# and finally fit a linear regression model.
pipe = Pipeline(Input)

# We will use the same features and data as in Question 7
# (X_multi and Y_multi), which have already been cleaned of NaNs.
if X_multi.empty or Y_multi.empty:
    print("Error: No data available to fit the pipeline (based on previous step).")
else:
    # Fit the pipeline to the data
    pipe.fit(X_multi, Y_multi)
    
    # Calculate and print the R^2
    # We use the pipeline's score() method on the same data
    r_squared_pipe = pipe.score(X_multi, Y_multi)
    print(f"The R-squared (R^2) value for the pipeline model is: {r_squared_pipe}")
```

# Module 5: Model Evaluation and Refinement


Import the necessary modules:


```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
print("done")
```

We will split the data into training and testing sets:


```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

### Question 9

Create and fit a Ridge regression object using the training data, set the regularization parameter to 0.1, and calculate the R^2 using the test data. Take a screenshot of your code and the value of the R^2. You will need to submit it for the final project.


```python
from sklearn.linear_model import Ridge
```

```python
# --- Question 9: Ridge Regression ---
print("\n--- Question 9: Ridge Regression ---")

from sklearn.linear_model import Ridge

# Create a Ridge regression object with regularization parameter 0.1
RidgeModel = Ridge(alpha=0.1)

# Fit the model using the training data
RidgeModel.fit(x_train, y_train)

# Calculate and print the R^2 using the test data
r_squared_ridge = RidgeModel.score(x_test, y_test)

print(f"The R-squared (R^2) value for the Ridge regression model (alpha=0.1) is: {r_squared_ridge}")
```

### Question 10

Perform a second order polynomial transform on both the training data and testing data. Create and fit a Ridge regression object using the training data, set the regularisation parameter to 0.1, and calculate the R^2 utilising the test data provided. Take a screenshot of your code and the R^2. You will need to submit it for the final project.


```python
# Create a PolynomialFeatures object of degree 2
pr = PolynomialFeatures(degree=2)

# Transform the training and testing data
x_train_poly = pr.fit_transform(x_train)
x_test_poly = pr.transform(x_test)

# Create a new Ridge regression object
RidgeModelPoly = Ridge(alpha=0.1)

# Fit the model using the polynomial-transformed training data
RidgeModelPoly.fit(x_train_poly, y_train)

# Calculate and print the R^2 using the polynomial-transformed test data
r_squared_poly = RidgeModelPoly.score(x_test_poly, y_test)
print(f"The R-squared (R^2) value for the 2nd-order Polynomial Ridge model (alpha=0.1) is: {r_squared_poly}")
```
